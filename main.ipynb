```python
import pandas as pd
import scipy.stats as stat
```

```python
df = pd.read_csv("out.csv")
```

```python
import math
```

```python
import matplotlib.pyplot as plt
```

```python
for vr in list(df.columns[2:]):
    corr = stat.pearsonr(df['time'].values, df[vr].values)
    spr = stat.spearmanr(df['time'].values, df[vr].values)
    if math.isnan(spr.correlation): continue
    print(f"{vr=}, {corr=}, {spr=}")
    plt.scatter(df[vr].values, df['time'].values)
    plt.xlabel(vr)
    plt.ylabel("time")
    plt.show()
```
